```python
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
from sklearn.datasets import make_moons, make_circles
import matplotlib.pyplot as plt
import pandas as pd, numpy as np, seaborn as sns, warnings as warn
warn.filterwarnings('ignore')
%matplotlib inline
```

```python
X,y = make_moons(n_samples = 1000, noise = 0.10, shuffle = True)
```

```python
std_scaler = StandardScaler()
X = std_scaler.fit_transform(X)
```

```python
plt.scatter(x = X[:,0], y = X[:,1], c = y)
```

```python
dbscan = DBSCAN(eps = 0.5, min_samples = 5)
```

```python
y_pred = dbscan.fit_predict(X)
```

```python
dbscan.fit(X)
```

```python
plt.scatter(x = X[:,0], y = X[:,1], c = y_pred)
```

```python
plt.scatter(x = X[:,0], y = X[:,1], c = dbscan.labels_)
```

## DBSCAN Clutering

```python
from sklearn.cluster import DBSCAN
from sklearn.datasets import make_moons
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.datasets import make_circles
```

```python
X, y = make_circles(n_samples=750, factor=0.3, noise=0.1)
```

```python
plt.scatter(X[:,0],X[:,1])
```

```python
from sklearn.cluster import DBSCAN
```

```python
dbcan=DBSCAN(eps=0.1)
```

```python
dbcan.fit_predict(X)
```

```python
dbcan.labels_
```

```python
plt.scatter(X[:,0],X[:,1],c=dbcan.labels_)
```

```python
plt.scatter(X[:,0],X[:,1],c=y)
```
