## Down Sampling

```python
clz_0 = pd.DataFrame({'data1': np.random.normal(loc = 0, scale = 1, size = 400), 
                      'data2': np.random.normal(loc=0, scale = 1, size=400), 'target': [0]*400})
clz_1 = pd.DataFrame({'data1': np.random.normal(loc = 0, scale = 1, size = 100), 
                      'data2': np.random.normal(loc=0, scale = 1, size=100), 'target': [1]*100})
df1 = pd.concat([clz_0, clz_1], ignore_index = 0)
```

```python
df1['target'].value_counts()
```

```python
df1_minor = df1[df1['target'] == 1]
df1_major = df1[df1['target'] == 0]
df1_minor_upsampled = resample(df1_minor, n_samples = len(df1_major), random_state = 42)
balanced_df1 = pd.concat([df1_major, df1_minor_upsampled], ignore_index = True)
```

```python
balanced_df1['target'].value_counts()
```

```python
df1_major_downsampled = resample(df1_major, n_samples = len(df1_minor), replace = False, random_state = 42)
balanced_df1 = pd.concat([df1_minor, df1_major_downsampled], ignore_index = True)
```

```python
balanced_df1['target'].value_counts()
```

```python
balanced_df1.duplicated().sum()
```

```python
df1_major_downsampled = resample(df1_major, n_samples = len(df1_minor), replace = True, random_state = 42)
balanced_df1 = pd.concat([df1_minor, df1_major_downsampled], ignore_index = True)
```
