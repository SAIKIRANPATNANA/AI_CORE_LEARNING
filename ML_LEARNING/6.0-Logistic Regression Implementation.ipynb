## Logistic Regression Implementation

```python
from sklearn.datasets import load_iris
```

```python
dataset=load_iris()
```

```python
print(dataset.DESCR)
```

```python
dataset.keys()
```

```python
import pandas as pd
import numpy as np
```

```python
df=pd.DataFrame(dataset.data,columns=dataset.feature_names)
```

```python
print(dataset.target)
```

```python
df['target']=dataset.target
```

```python
df.head()
```

```python
df['target'].unique()
```

```python
df[df['target']!=2]
```

```python
df_copy=df[df['target']!=2]
```

```python
df_copy['target'].unique()
```

```python
## Independent and dependent features
X=df_copy.iloc[:,:-1]
y=df_copy.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
classification=LogisticRegression(max_iter=200)
```

```python
classification
```

```python
classification.fit(X_train,y_train)
```

```python
y_pred=classification.predict(X_test)
```

```python
y_pred
```

## Confusion Metrics,Accuracy,Classification Report

```python
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
```

```python
print(confusion_matrix(y_test,y_pred))
```

```python
print(accuracy_score(y_test,y_pred))
```

```python
print(classification_report(y_test,y_pred))
```

## Cross Validation

```python
from sklearn.model_selection import KFold
```

```python
cv=KFold(n_splits=5)
```

```python
cv
```

```python
from sklearn.model_selection import cross_val_score
```

```python
## these are the scores all the cross validation
scores=cross_val_score(classification,X_train,y_train,scoring='accuracy',cv=cv)
```

```python
np.mean(scores)
```
