## Feature Scaling

1. Standardization
2. Normalization-Min Max Scaler

```python
import seaborn as sns
import warnings as warn 
warn.filterwarnings('ignore')
```

```python
df=sns.load_dataset('tips')
```

```python
df.head()
```

```python
import numpy as np
mean = np.mean(df['total_bill'])
std = np.std(df['total_bill'])
standardized_data = list()
for data in df['total_bill']:
    standardized_data.append((data-mean)/std)
```

```python
sns.histplot(df['total_bill'], kde = True)
```

```python
sns.histplot(standardized_data, kde = True)
```

```python
from sklearn.preprocessing import StandardScaler
std = StandardScaler()
standardized_data = std.fit_transform(df[['total_bill']])
```

```python
sns.histplot(standardized_data)
```

## Normalization--Min Max Scaler

```python
min_max_data = list()
min = min(df['total_bill'])
max = max(df['total_bill'])
for data in df['total_bill']:
    min_max_data.append((data-min)/(max-min))
```

```python
sns.histplot(df['total_bill'])
```

```python
sns.histplot(min_max_data)
```

```python
from sklearn.preprocessing import MinMaxScaler
min_max = MinMaxScaler()
min_max_data = min_max.fit_transform(df[['total_bill']])
```

```python
sns.histplot(min_max_data)
```

```python
pd.DataFrame(min_max.fit_transform(df[['total_bill', 'tip']]), columns = ['total_bill', 'tip'])
```
